```python
import pandas as pd
import matplotlib.pyplot as plt 
pt = pd.read_csv("Sales.csv")
x =pt['Units Sold']
y =pt['Unit Price']
plt.plot(x, y,marker='*')
#scatter graph
# plt.scatter(x,y, marker='*')
plt.title("Unit sold vs price")
plt.xlabel("Units Sold")
plt.ylabel("Unit Price")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pt = pd.read_csv("Sales.csv")
priority_counts = pt['Order Priority'].value_counts()
plt.pie(priority_counts, labels=priority_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Order Priority Distribution')
plt.axis('equal')  
plt.show()
```
